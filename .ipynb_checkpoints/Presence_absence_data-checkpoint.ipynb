```python
#This converts the pickled data files into one long list made up of tuples with location and species name.
#Each tuple represents the presence of that species at that location.
#This step removes site without observations, which is good, I think.
aggregated_data = []
species_codes = []
for prefix in prefixes:
    file_name="data/"+prefix+"_taxon_data.pkl"
    region_taxon_data = pickle.load(open(file_name, "rb" ))
    for i in xrange(10):
        for j in xrange(10):
            site = prefix+str(i)+str(j)
            for k in xrange(len(region_taxon_data[site])):
                if region_taxon_data[site][k]['taxon']['commonName']:
                    print '!'
                else:
                    print '?'
                aggregated_data.append([site, 
                                        region_taxon_data[site][k]['taxon']['name'],
                                        region_taxon_data[site][k]['taxon']['taxonVersionKey'],
                                        region_taxon_data[site][k]['taxon']['commonName']
                                       ])
    print prefix
```
